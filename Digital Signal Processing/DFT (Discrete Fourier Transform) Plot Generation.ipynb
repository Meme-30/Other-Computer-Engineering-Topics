```python
# Sampling freq
fs = 2000
# Freq of analog signal
f1 = 304
# Number of samples acquired
N = 100

# DO NOT EDIT BELOW THIS LINE
```

```python
Ts = 1 / fs

# Draw hi-res y(t)
nval = np.arange(0, N * 20)
tval = np.arange(0, N * 20) * (Ts / 20)
yval = np.sin(2 * np.pi * f1 * tval)

plt.figure(figsize=(18, 6), dpi=80)
plt.plot(tval, yval)
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)

# Draw samples
nval = np.arange(0, N)
tval = np.arange(0, N) * Ts
xval = np.sin(2 * np.pi * f1 * tval)

plt.stem(tval, xval)
plt.grid(True)

plt.plot(nval * Ts, xval, 'ro')
```

```python
X = np.fft.fft(xval)
freq = np.fft.fftfreq(len(X)) * fs
plt.figure(figsize=(16,12))
plt.stem(freq, np.abs(X)**2)
plt.plot(freq, np.abs(X)**2,'r-')
plt.grid()
#plt.xlim([280,320])
plt.title('DFT plot')
```

```python
X = np.fft.fft(xval)
freq = np.fft.fftfreq(len(X)) * fs
plt.figure(figsize=(16,12))
plt.stem(freq, np.abs(X)**2)
plt.plot(freq, np.abs(X)**2,'r-')
plt.grid()
plt.xlim([200,390])
plt.title('DFT plot (zoom)')
```
